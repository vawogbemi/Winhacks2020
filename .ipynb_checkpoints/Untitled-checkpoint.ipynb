## Getting Started

For this challenge we were provided with 3 datasets(Dorchester Road and Huron Church Road.csv, Malden Road and Huron Church Road.csv, Totten Street and Huron Church Road.csv); each one corresponding to a busy intersection in Windsor. The datasets have columns which represent start time and turns (right, left, thru, u-turn, peds_cw, peds_ccw) for each cardinal direction (north, east, south, west), in total there are 25 columns. Each row represents 15 min of time and each column records the number of occurances within 15 mins intervals.Lastly there are multiple tables in each dataset there is a main one which totals occurances for every vehicle and several seperate tables for individual vehicles, each table is 5760 rows long.
